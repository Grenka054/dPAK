```python
class Item:
    def __init__(self, count=3, max_count=16):
        self._count = count
        self._max_count = max_count

    def update_count(self, val):
        if val <= self._max_count:
            self._count = val
            return True
        else:
            return False
        
    # перенесенные функции 
    def __add__(self, num):
        """ Сложение с числом """
        return self._count + num

    def __mul__(self, num):
        """ Умножение на число """
        return self._count * num

    def __lt__(self, num):
        """ Сравнение меньше """
        return self._count < num

    def __len__(self):
        """ Получение длины объекта """
        return self._count

    @property
    def count(self):
        return self._count
```

```python
item = Item(2)
print(len(item))
print(item + 3)
print(item * 3)
print(item < 3)
```

```python
class Item:
    def __init__(self, count=3, max_count=16):
        self._count = count
        self._max_count = max_count

    def update_count(self, val):
        if 0 <= val <= self._max_count:
            self._count = val
            return True
        else:
            return False

    def __add__(self, num):
        """ Сложение с числом """
        return self._count + num

    def __mul__(self, num):
        """ Умножение на число """
        return self._count * num

    def __lt__(self, num):
        """ Сравнение меньше """
        return self._count < num

    def __len__(self):
        """ Получение длины объекта """
        return self._count
    # и еще операции
    def __sub__(self, num):
        """ Вычитание числа """
        return self._count - num

    def __gt__(self, num):
        """ Сравнение больше """
        return self._count > num

    def __ge__(self, num):
        """ Сравнение не меньше """
        return self._count >= num

    def __le__(self, num):
        """ Сравнение не больше """
        return self._count <= num

    def __eq__(self, num):
        """ Проверка на эквивалентность """
        return self._count == num

    def __iadd__(self, num):
        """ Сложение с числом и присваивание результата """
        self.update_count(self._count + num)
        return self

    def __isub__(self, num):
        """ Вычитание числа и присваивание результата """
        self.update_count(self._count - num)
        return self

    def __imul__(self, num):
        """ Умножение на число и присваивание результата """
        self.update_count(self._count * num)
        return self

    @property
    def count(self):
        return self._count
```

```python
item = Item(2)
print(item - 3)
print(item > 3)
print(item <= 3)
print(item >= 3)
print(item == 3)
print(len(item))
item += 5
print(len(item))
item *= 5
print(len(item))
item -= 1
print(len(item))
```

```python
class Food(Item):
    def __init__(self, saturation, **kwargs):
        super().__init__(**kwargs)
        self._saturation = saturation

    @property
    def eatable(self):
        return self._saturation > 0


class Fruit(Item):
    def __init__(self, ripe=True, **kwargs):
        super().__init__(**kwargs)
        self._ripe = ripe


class Apple(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color

    @property
    def color(self):
        return self._color

    @property
    def eatable(self):
        return super().eatable and self._ripe

# свои классы
class Banana(Fruit, Food):
    def __init__(self, ripe, count=3, max_count=50, color='yellow', saturation=11):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color

    @property
    def color(self):
        return self._color

    @property
    def eatable(self):
        return super().eatable and self._ripe


class Pomelo(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=3, color='olive', saturation=32):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color

    @property
    def color(self):
        return self._color

    @property
    def eatable(self):
        return super().eatable and self._ripe


class Tomato(Food):
    def __init__(self, count=3, max_count=50, color='red', saturation=5, **kwargs):
        super().__init__(saturation=saturation, count=count, max_count=max_count, **kwargs)
        self._color = color

    @property
    def color(self):
        return self._color


class Meat(Food):
    def __init__(self, fresh, weight=1, max_weight=50, saturation=5, **kwargs):
        super().__init__(saturation=saturation, count=weight, max_count=max_weight, **kwargs)
        self._fresh = fresh

    @property
    def eatable(self):
        return super().eatable and self._fresh
```

```python
meat = Meat(True, 2)
print(meat.eatable)

banana = Banana(True, 1)
print(banana.eatable)

tomato = Tomato(saturation=0)
print(tomato.eatable)

pomelo = Pomelo(True, 2)
print(pomelo.eatable)
```

```python
class Apple(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color

    # перегрузка вывода
    def __str__(self):
        return f"(Apple, {self.count})"

    @property
    def color(self):
        return self._color

    @property
    def eatable(self):
        return super().eatable and self._ripe


class Banana(Fruit, Food):
    def __init__(self, ripe, count=3, max_count=50, color='yellow', saturation=11):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color

    def __str__(self):
        return f"(Banana, {self.count})"

    @property
    def color(self):
        return self._color

    @property
    def eatable(self):
        return super().eatable and self._ripe


class Pomelo(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=3, color='olive', saturation=32):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color

    def __str__(self):
        return f"(Pomelo, {self.count})"

    @property
    def color(self):
        return self._color

    @property
    def eatable(self):
        return super().eatable and self._ripe


class Tomato(Food):
    def __init__(self, count=3, max_count=50, color='red', saturation=5, **kwargs):
        super().__init__(saturation=saturation, count=count, max_count=max_count, **kwargs)
        self._color = color

    def __str__(self):
        return f"(Tomato, {self.count})"

    @property
    def color(self):
        return self._color


class Meat(Food):
    def __init__(self, fresh, weight=1, max_weight=50, saturation=5, **kwargs):
        super().__init__(saturation=saturation, count=weight, max_count=max_weight, **kwargs)
        self._fresh = fresh

    def __str__(self):
        return f"(Meat, {self.count})"

    @property
    def eatable(self):
        return super().eatable and self._fresh


class Inventory:
    def __init__(self, size=10):
        self._items = []
        for i in range(size):
            self._items.append(None)
        self._size = size

    def __len__(self):
        return self._size

    def update_list(self, index, item):
        self._items[index] = item
        return True

    def __str__(self):
        str_output = ""
        for item in self._items:
            str_output += "%s " % item
        return str_output

    def pop(self, index):
        if self._items[index] is None:
            return False
        self._items[index] -= 1
        if self._items[index] <= 0:
            self._items[index] = None
        return True

    @property
    def items(self):
        return self._items
```

```python
meat = Meat(True, 2)
banana = Banana(True, 1)
inventory = Inventory(4)
inventory.update_list(0, meat)
inventory.update_list(3, banana)
print(inventory)
inventory.pop(0)
print(inventory)
inventory.pop(0)
print(inventory)
inventory.pop(3)
print(inventory)
inventory.pop(0)
print(inventory)
```

```python
class Queue:
    def __init__(self):
        self._list = []

    def push_back(self, val):
        self._list.append(val)

    def pop(self):
        """ Извлечение элемента """
        if self._list:
            removed = self._list.pop(0)
            return removed
        else:
            return None

    def __str__(self):
        str_output = ""
        for elem in self._list:
            str_output += "%s " % elem
        return str_output

    def copy_list(self):
        return self._list.copy()
```

```python
queue = Queue()
queue.push_back(4)
queue.push_back(6)
print(queue)
queue.pop()
queue.push_back(8)
print(queue)
queue.pop()
print(queue)
queue.pop()
print(queue)
queue.push_back(4)
queue.push_back(6)
queue.push_back(8)
print(queue)
a = queue.copy_list()
a.append(5)
print(a)
print(queue)
```
